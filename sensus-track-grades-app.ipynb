```python
class SensuTrack:
    def __init__(self, id):
        self.id = id
        self.percentPerTrack = 0
        self.percentPerTrackTimesGrade = 0
        self.grades_to_values = {"A":1, "B":2, "C":3, "D":4}
        self.values_to_grades = {1:"A", 2:"B", 3:"C", 4:"D"}

    def grade(self):
        return self.values_to_grades[round(self.percentPerTrackTimesGrade / self.percentPerTrack)]
    
    def gradeAsNum(self):
        return self.percentPerTrackTimesGrade / self.percentPerTrack
    
    def addPercentPerTrack(self, num, grade):
        self.percentPerTrack += num
        self.percentPerTrackTimesGrade += num * self.grades_to_values[grade]
```

This next code does something cool

```python
import pandas as pd

file_path = "sensus-track-data.xlsx"
sensus_tracks = pd.read_excel(file_path)
print(sensus_tracks.head())
```

```python
sensus_tracks_grades = {}

for index, row in sensus_tracks.iterrows():
    grade = row["Grade"]
    
    if (not pd.isna(grade) and grade != "Commercial" and grade != "Industrial"):
        
        if not pd.isna(row["Associated Census Tract ID"]):
            current_sensus_track_id = str(int(row["Associated Census Tract ID"]))

        if (current_sensus_track_id not in sensus_tracks_grades):
            sensus_tracks_grades[current_sensus_track_id] = SensuTrack(current_sensus_track_id)
            
            
        current_sensus_track = sensus_tracks_grades[current_sensus_track_id]
        current_sensus_track.addPercentPerTrack(row["Percent Per Tract"], grade)
```

```python
for key, value in sensus_tracks_grades.items():
    print(value.id, ": ", round(value.percentPerTrack,3), "\t", round(value.percentPerTrackTimesGrade, 3), "\t", value.grade(), "\t", value.gradeAsNum())
```
